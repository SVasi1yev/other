```python
import pandas as pd
import numpy as np
```

```python
manual = pd.read_csv('manual_z5_full.csv')
```

```python
manual
```

```python
objid = list(manual['OBJID'])
manual['OBJID'] = list(map(lambda x: x[2:-1], objid))
```

```python
manual
```

```python
plates = list(map(lambda x: x.split('-')[0], manual['OBJID']))
mjds = list(map(lambda x: x.split('-')[1], manual['OBJID']))
fibers = list(map(lambda x: x.split('-')[2], manual['OBJID']))
```

```python
manual.insert(1, 'plate', plates)
manual.insert(2, 'mjd', mjds)
manual.insert(3, 'fiber', fibers)
```

```python
manual
```

```python
links = []
for i in range(len(plates)):
    links.append(f'http://skyserver.sdss.org/dr17/VisualTools/quickobj?plate={plates[i]}&mjd={mjds[i]}&fiber={fibers[i]}')
```

```python
manual['link'] = links
```

```python
manual.insert(0, 'idx', range(manual.shape[0]))
```

```python
manual
```

```python
from bs4 import BeautifulSoup
import requests
from PIL import Image
from io import BytesIO
from tqdm import tqdm
```

```python
for i, url in tqdm(enumerate(finish['link']), total=finish.shape[0]):
    if i < 94:
        continue
    page = requests.get(url)
    soup = BeautifulSoup(page.text, "html.parser")
    img_url = 'http://skyserver.sdss.org/' + soup.find_all('img', class_='img-thumbnail')[1]['src']
    response = requests.get(img_url)
    img = Image.open(BytesIO(response.content)).convert('RGB').save(f'manual_z5_full_imgs/{str(i).zfill(3)}.jpg')
```
