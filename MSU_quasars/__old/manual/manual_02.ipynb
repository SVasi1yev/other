```python
df_01 = pd.read_csv('manual_z5_full_20230130.csv')
df_01
```

```python
df_02 = pd.read_csv('dr18_z5.csv')[['plate', 'mjd', 'fiberID', 'ra', 'dec', 'z']]
df_02.rename(columns={'fiberID': 'fiber'}, inplace=True)

objid = []
for i in range(df_02.shape[0]):
    objid.append(f'{df_02.loc[i, "plate"]}-{df_02.loc[i, "mjd"]}-{str(df_02.loc[i, "fiber"]).zfill(4)}')
df_02['OBJID'] = objid
df_02
```

```python
# df_03 = df_02.merge(df_01, how='left', on=['OBJID'])
df_03 = df_02.merge(df_01, how='left', on=['plate', 'mjd', 'fiber'])
# df_03.rename(columns={
#     'plate_x': 'plate', 'mjd_x': 'mjd', 'fiber_x': 'fiber'
# }, inplace=True)
df_03.rename(columns={'OBJID_x': 'OBJID'}, inplace=True)
df_03
```

```python
(~df_03['Z_SYS'].isna()).sum()
```

```python
df_04 = df_03[['plate',
 'mjd',
 'fiber',
 'ra',
 'dec',
 'z',
 'OBJID',
 'Comment',
 'Interesting_Score_max10',
 'Owner',
 'Vizzspec',
 'Vizclass',
 'Z_SYS',
 'rf_photo_z(2)',
 'nn_photo_z(3)',
 'link',
 'target',
 'boost_preds',
 'nn_preds']]
df_04
```

```python
df_05 = df_04.copy()
```

```python
df_05['index'] = range(df_05.shape[0])
plates = list(map(lambda x: x.split('-')[0], df_05['OBJID']))
mjds = list(map(lambda x: x.split('-')[1], df_05['OBJID']))
fibers = list(map(lambda x: x.split('-')[2], df_05['OBJID']))
links = []
for i in range(len(plates)):
    links.append(f'http://skyserver.sdss.org/dr17/VisualTools/quickobj?plate={plates[i]}&mjd={mjds[i]}&fiber={fibers[i]}')
df_05['link'] = links    
```
